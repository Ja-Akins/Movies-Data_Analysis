# Data Reduction and Preparation

## Pick columns that are necessary for the analysis.

```python
movies_data = movies_data[['budget','genres','original_title','overview','popularity','production_companies',
         'production_countries', 'release_date','revenue','runtime','vote_average','vote_count','cast','crew']]
```

```python
# Create a copy to prevent the original data from being unintentionally modified by the operations I perform
movies_data = movies_data.copy()
```

```python
# Remove entries with zero budget or revenue
movies_data = movies_data[(movies_data['budget'] > 0) & (movies_data['revenue'] > 0)]
movies_data.dropna(subset=['runtime', 'release_date'], inplace=True)
```

```python
# Date conversion
movies_data['release_date'] = pd.to_datetime(movies_data['release_date'], errors='coerce')
movies_data['year'] = movies_data['release_date'].dt.year
```

```python
# Extract genres
movies_data['genres'] = movies_data['genres'].apply(ast.literal_eval).apply(
    lambda x: [d['name'] for d in x] if isinstance(x, list) else []
)   
```

```python
# Extract directors from crew
movies_data['crew'] = movies_data['crew'].apply(ast.literal_eval)
movies_data['director'] = movies_data['crew'].apply(
    lambda x: next((d['name'] for d in x if d['job'] == 'Director'), None)
)
```

```python
movies_data['director']
```

```python
# Extract main actors from cast (first 3)
movies_data['cast'] = movies_data['cast'].apply(ast.literal_eval)
movies_data['main_actors'] = movies_data['cast'].apply(
    lambda x: [d['name'] for d in x[:3]] if isinstance(x, list) else []
)
```

```python
movies_data['main_actors']
```

```python
# Financial metrics
movies_data['profit'] = movies_data['revenue'] - movies_data['budget']
movies_data['ROI'] = np.where(movies_data['budget'] != 0, movies_data['profit'] / movies_data['budget'], 0)
```
